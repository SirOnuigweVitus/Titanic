```python
dt = df.drop(["name", "cabin", "TickNum", "MedBoat"], axis=1)
dt.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labE = LabelEncoder()
X=dt
X["embarked"] = labE.fit_transform(X["embarked"])
X
```

```python
from sklearn.preprocessing import OneHotEncoder
OneH = OneHotEncoder(categorical_features=[-3])
X = OneH.fit_transform(X).toarray()
X
```

```python
X = X[:, 1:]
X
```

```python
X = dt[["traveller_ID", "ticket_class", "sex", "age", "Siblings_spouses", "Parchil", "fare", "embarked"]].values
X
```

```python
y = dt.Survived
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30,random_state =42)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
ln = LogisticRegression()
model = ln.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print (accuracy_score(y_test, model.predict(X_test)))
```
